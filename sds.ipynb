```python
keys = ['one', 'two', 'three']
values = [1,2,3]
def join_dict(d1: list, d2: list) -> dict:
    slovar = {}
    for i in range(len(d1)):
        slovar[d1[i]] = d2[i]
    return slovar
print(join_dict(keys, values))

def comprehensions(d1: list, d2: list) -> dict:
    slovar = {d1[i]:d2[i] for i in range(len(d1))}
    return slovar

print(comprehensions(keys, values))
```
